# It can't work on Non Linear data. 
- Solution is Multi Layer Perceptron

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1'] = [0,1,1,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
or_data
```

```python
sns.scatterplot(data = or_data, x=or_data['input1'], y=or_data['input1'], hue = or_data['output'])
```

```python
and_data['input1'] = [1,0,1,0]
and_data['input2'] = [1,0,0,1]
and_data['output'] = [1,0,0,0]
and_data
```

```python
xor_data['input1'] = [1,0,1,0]
xor_data['input2'] = [1,0,0,1]
xor_data['output'] = [0,0,1,1]
xor_data
```

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
clf1.coef_ # value of w1 and w2
```

```python
clf1.intercept_ # value of b
```

```python
x=np.linspace(-1,1,5)
y=-x+1
```

```python
plt.plot(x,y)
sns.scatterplot(data = and_data, x = and_data['input1'], y = and_data['input2'],hue=and_data['output'],s=200)
```

```python
clf2.coef_
```

```python
clf2.intercept_
```

```python
x1=np.linspace(-1,1,5)
y1=-x+0.5
```

```python
plt.plot(x1,y1)
sns.scatterplot(data = or_data , x = or_data['input1'], y = or_data['input2'],hue=or_data['output'],s=200)
```

```python
print(clf3.coef_)

print(clf3.intercept_)
```

```python
x2=np.linspace(-1,1,5)
y2=-x+0.5
```

```python
plt.plot(x2,y2)
sns.scatterplot(data = xor_data , x = xor_data['input1'], y = xor_data['input2'],hue=xor_data['output'],s=200)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values, clf=clf2, legend=2)

# a decision region can be formedd in OR data case to classify the points
```

```python
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values, clf=clf3, legend=2)

# there is no decision region drawn to separate XOR data which is non linear i.e perceptron is unable to classify non linear data
```
