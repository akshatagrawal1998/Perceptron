```python
# Independent and Dependent Features 

x = df.iloc[:,0:2]
y = df.iloc[:,-1]
print(x)
```

```python
print(y)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x,y)
print("Value of A and B in Ax+By+c=0 : ",p.coef_)
print("Value of c in Ax+By+c=0 : ",p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=p, legend=2) # Plot decision regions of a classifier.
```
